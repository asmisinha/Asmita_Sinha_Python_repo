```python
def display_menu():
    print("\n--- Library Inventory System ---")
    print("1. Enter book values")
    print("2. Display all book records")
    print("3. Total number of books and average value")
    print("4. Highest and lowest book value")
    print("5. Display premium books")
    print("6. Search for a book value")
    print("7. Count occurrences of a book value")
    print("8. Remove duplicate book records")
    print("9. Display books in reverse order")
    print("10. Find index of a book value")
    print("0. Exit")


books = () 

while True:
    display_menu()
    choice = int(input("Enter your choice: "))

    if choice == 1:
        n = int(input("How many books do you want to enter? "))
        temp_list = []
        for i in range(n):
            value = float(input(f"Enter value of book {i + 1}: "))
            temp_list.append(value)
        books = books + tuple(temp_list)
        print("Books added successfully.")

    elif choice == 2:
        print("Book Records:", books)

    elif choice == 3:
        if books:
            total_books = len(books)
            average = sum(books) / total_books
            print("Total number of books:", total_books)
            print("Average book value:", average)
        else:
            print("No books available.")

    elif choice == 4:
        if books:
            print("Highest book value:", max(books))
            print("Lowest book value:", min(books))
        else:
            print("No books available.")

    elif choice == 5:
        limit = float(input("Enter premium book limit: "))
        premium_books = tuple(b for b in books if b >= limit)
        print("Premium Books:", premium_books)

    elif choice == 6:
        search_value = float(input("Enter book value to search: "))
        if search_value in books:
            print("Book value exists in the system.")
        else:
            print("Book value not found.")

    elif choice == 7:
        count_value = float(input("Enter book value to count: "))
        print("Occurrences:", books.count(count_value))

    elif choice == 8:
        unique_books = tuple(set(books))
        books = unique_books
        print("Duplicates removed.")
        print("Updated book records:", books)

    elif choice == 9:
        print("Books in reverse order:", books[::-1])

    elif choice == 10:
        index_value = float(input("Enter book value to find index: "))
        if index_value in books:
            print("Index position:", books.index(index_value))
        else:
            print("Book value not found.")

    elif choice == 0:
        print("Exiting Library Inventory System.")
        break

    else:
        print("Invalid choice. Please try again.")
```
