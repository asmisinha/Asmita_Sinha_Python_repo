```python
# Bank Account Management System using Dictionary

accounts = {}

def create_account():
    acc_no = int(input("Enter account number: "))
    if acc_no in accounts:
        print("Account already exists.")
    else:
        balance = float(input("Enter initial balance: "))
        accounts[acc_no] = balance
        print("Account created successfully.")

def display_accounts():
    if not accounts:
        print("No accounts available.")
    else:
        print("\nAccount Number : Balance")
        for acc, bal in accounts.items():
            print(acc, ":", bal)

def total_and_average_balance():
    if not accounts:
        print("No accounts available.")
    else:
        total = sum(accounts.values())
        average = total / len(accounts)
        print("Total Balance:", total)
        print("Average Balance:", average)

def highest_and_lowest_balance():
    if not accounts:
        print("No accounts available.")
    else:
        max_acc = max(accounts, key=accounts.get)
        min_acc = min(accounts, key=accounts.get)
        print("Highest Balance:", accounts[max_acc], "in Account:", max_acc)
        print("Lowest Balance:", accounts[min_acc], "in Account:", min_acc)

def high_value_accounts():
    limit = float(input("Enter balance limit: "))
    found = False
    for acc, bal in accounts.items():
        if bal > limit:
            print(acc, ":", bal)
            found = True
    if not found:
        print("No accounts above the given limit.")

def search_account():
    acc_no = int(input("Enter account number to search: "))
    if acc_no in accounts:
        print("Account found. Balance:", accounts[acc_no])
    else:
        print("Account does not exist.")

def update_balance():
    acc_no = int(input("Enter account number: "))
    if acc_no in accounts:
        amount = float(input("Enter amount (+deposit / -withdraw): "))
        accounts[acc_no] += amount
        print("Updated Balance:", accounts[acc_no])
    else:
        print("Account not found.")

def remove_account():
    acc_no = int(input("Enter account number to remove: "))
    if acc_no in accounts:
        del accounts[acc_no]
        print("Account removed successfully.")
    else:
        print("Account not found.")

def display_separately():
    print("Account Numbers:", list(accounts.keys()))
    print("Balances:", list(accounts.values()))

def sort_accounts():
    if not accounts:
        print("No accounts available.")
    else:
        print("\nAccounts sorted by account number:")
        for acc in sorted(accounts):
            print(acc, ":", accounts[acc])

# Menu-driven program
while True:
    print("\n--- Bank Account Management System ---")
    print("1. Create Account")
    print("2. Display All Accounts")
    print("3. Total and Average Balance")
    print("4. Highest and Lowest Balance")
    print("5. High-Value Accounts")
    print("6. Search Account")
    print("7. Update Balance (Deposit/Withdraw)")
    print("8. Remove Account")
    print("9. Display Account Numbers and Balances Separately")
    print("10. Sort Accounts by Account Number")
    print("11. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        create_account()
    elif choice == 2:
        display_accounts()
    elif choice == 3:
        total_and_average_balance()
    elif choice == 4:
        highest_and_lowest_balance()
    elif choice == 5:
        high_value_accounts()
    elif choice == 6:
        search_account()
    elif choice == 7:
        update_balance()
    elif choice == 8:
        remove_account()
    elif choice == 9:
        display_separately()
    elif choice == 10:
        sort_accounts()
    elif choice == 11:
        print("Exiting program...")
        break
    else:
        print("Invalid choice. Please try again.")
```
