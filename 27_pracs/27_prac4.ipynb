```python
print("Asmita Sinha, 24204A0027")

student_marks = []

pass_mark = 40
distinction_mark = 75

while True:
    print("\n--- STUDENT MANAGEMENT SYSTEM ---")
    print("1. Enter student marks")
    print("2. Display all student marks")
    print("3. Display total and average marks")
    print("4. Display highest and lowest marks")
    print("5. Display passed and failed students")
    print("6. Display distinction students")
    print("7. Remove duplicate marks")
    print("8. Replace failed marks with grace marks")
    print("9. Search for a student mark")
    print("10. Sort marks and display reverse order")
    print("11. Remove a student record")
    print("12. Find position of a student mark")
    print("0. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        mark = int(input("Enter student mark: "))
        student_marks.append(mark)
        print("Mark added successfully")

    elif choice == 2:
        print("Student Marks:", student_marks)

    elif choice == 3:
        total = sum(student_marks)
        average = total / len(student_marks) if student_marks else 0
        print("Total Marks:", total)
        print("Average Marks:", average)

    elif choice == 4:
        if student_marks:
            print("Highest Mark:", max(student_marks))
            print("Lowest Mark:", min(student_marks))
        else:
            print("No records available")

    elif choice == 5:
        print("Passed Students:")
        for mark in student_marks:
            if mark >= pass_mark:
                print(mark)

        print("Failed Students:")
        for mark in student_marks:
            if mark < pass_mark:
                print(mark)

    elif choice == 6:
        print("Distinction Students:")
        for mark in student_marks:
            if mark >= distinction_mark:
                print(mark)

    elif choice == 7:
        student_marks = list(set(student_marks))
        print("Duplicate marks removed")

    elif choice == 8:
        for i in range(len(student_marks)):
            if student_marks[i] < pass_mark:
                student_marks[i] = pass_mark
        print("Grace marks applied")

    elif choice == 9:
        search_mark = int(input("Enter mark to search: "))
        if search_mark in student_marks:
            print("Mark found")
        else:
            print("Mark not found")

    elif choice == 10:
        student_marks.sort()
        print("Sorted Marks:", student_marks)
        print("Reverse Order:", student_marks[::-1])

    elif choice == 11:
        mark = int(input("Enter mark to remove: "))
        if mark in student_marks:
            student_marks.remove(mark)
            print("Student record removed")
        else:
            print("Mark not found")

    elif choice == 12:
        search_mark = int(input("Enter marks to find position: "))
        if search_mark in student_marks:
            print("Position (Index):", student_marks.index(search_mark))
        else:
            print("Mark not found")

    elif choice == 0:
        print("Exiting Program")
        break

    else:
        print("Invalid choice")
```
