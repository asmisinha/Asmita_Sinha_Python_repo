2	Calculate equivalent resistance of resistors connected in series and parallel

```python
R1 = float(input("Enter resistance R1 (Ohms): "))
R2 = float(input("Enter resistance R2 (Ohms): "))
R3 = float(input("Enter resistance R3 (Ohms): "))

# Series resistance
R_series = R1 + R2 + R3

# Parallel resistance
R_parallel = 1 / ((1 / R1) + (1 / R2) + (1 / R3))

print("Equivalent Resistance in Series =", R_series, "Ohms")
print("Equivalent Resistance in Parallel =", R_parallel, "Ohms")
```

3	Calculate voltage using Ohm’s Law

```python
R = float(input("Enter resistance R (Ohms): "))
I = float(input("Enter Current I (Ampere): "))

Voltage = I * R

print("Voltage = ",Voltage, "V")
```

4	Check whether the entered frequency is audio or radio frequency

```python
frequency = float(input("Enter frequency in Hz: "))

if 20 <= frequency <= 20000:
    print("The entered frequency is an Audio Frequency")
elif frequency > 20000:
    print("The entered frequency is a Radio Frequency")
else:
    print("Frequency is below audible range")
```

5	Display various radio frequency bands using if–elif ladder

```python
freq = float(input("Enter frequency in Hz: "))

if 30000 <=freq < 300000:
    print("Very Low Frequency")
elif 300000 <= freq < 3000000:
    print("Low Frequency")
elif 3000000 <= freq > 3000000 :
    print("High Frequency")
elif 3000000 <= freq > 30000000 :
    print ("Very High frequency")
else:
    print("Frequency not in Standard RF range")
```

6	Display resistor color code using match-case

```python
color = input("Enter resistor color: ").lower()

match color:
    case "black":
        print("Value = 0")
    case "brown":
        print("Value = 1")
    case "red":
        print("Value = 2")
    case "orange":
        print("Value = 3")
    case "yellow":
        print("Value = 4")
    case "green":
        print("Value = 5")
    case "blue":
        print("Value = 6")
    case "violet":
        print("Value = 7")
    case "grey":
        print("Value = 8")
    case "white":
        print("Value = 9")
    case _:
        print("Invalid color entered")
```
