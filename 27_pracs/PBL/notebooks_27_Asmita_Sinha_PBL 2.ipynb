Problem Statements / Goals

1. List Operations
Develop a Python program to create, access, update, and delete student records using a List data structure.

```python
students = []

students.append("Alice")
students.append("Bob")

print("Students:", students)
print("First student:", students[0])

students[1] = "Charlie"

students.remove("Alice")

print("Updated Students:", students)
```

2. Tuple Operations
Write a Python program to store fixed laboratory information using a Tuple and perform access and modification operations.

```python
lab_info = ("Physics Lab", 101, "Dr. Jons")

print("Lab Name:", lab_info[0])
print("Lab Number:", lab_info[1])

lab_info = ("Chemistry Lab", lab_info[1], lab_info[2])
print("Updated Lab Info:", lab_info)
```

3. Set Operations
Develop a Python program to manage unique laboratory equipment records using a Set and perform add and remove operations.

```python
equipment = {"Microscope", "Beaker", "Test Tube"}
equipment.add("Bunsen Burner")
equipment.remove("Beaker")

print("Laboratory Equipment:", equipment)
```

4. Dictionary Operations
Write a Python program to store and manage student details (Roll No, Name, Marks) using a Dictionary, including create, access, update, delete, and traversal operations.


```python
students = {
    1: {"Name": "Alice", "Marks": 85},
    2: {"Name": "Bob", "Marks": 90}
}

print("Student 1:", students[1])

students[2]["Marks"] = 95

del students[1]

for roll_no, details in students.items():
    print(roll_no, details)
```

5. Student Record Management
Design a Python program using appropriate data structures to maintain and display multiple student records.

```python
students = [
    {"Roll": 1, "Name": "Alice", "Marks": 85},
    {"Roll": 2, "Name": "Bob", "Marks": 90}
]

# Display records
for student in students:
    print(student)
```

6. Laboratory Record Handling
Develop a Python program to manage laboratory resources and availability using Python data structures.

```python
lab_resources = {
    "Microscope": 5,
    "Beaker": 20,
    "Test Tube": 50
}

# Check availability
for item, quantity in lab_resources.items():
    print(item, "available:", quantity)

# Update availability
lab_resources["Beaker"] -= 5
print("Updated Resources:", lab_resources)
```

7. Data Traversal
Write a Python program to traverse and display all stored student and laboratory records efficiently.



```python
students = ["Alice", "Bob", "Charlie"]
lab_equipment = {"Microscope", "Beaker", "Test Tube"}

print("Students:")
for student in students:
    print(student)

print("\nLab Equipment:")
for item in lab_equipment:
    print(item)
```
