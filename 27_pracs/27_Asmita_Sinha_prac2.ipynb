Write a Python program that checks if a resistor value is within a tolerance range (e.g., 5% tolerance for a 100Ω resistor) and prints whether it is within tolerance or not.

```python
nominal_value = 100
tolerance_percent = 5
measured_value = float(input("Enter the measured resistor value (in ohms): "))
tolerance = (tolerance_percent / 100) * nominal_value
lower_limit = nominal_value - tolerance
upper_limit = nominal_value + tolerance
if lower_limit <= measured_value <= upper_limit:
    print("The resistor is WITHIN tolerance.")
else:
    print("The resistor is OUT OF tolerance.")
print(f"Allowed range: {lower_limit}Ω to {upper_limit}Ω")
```
