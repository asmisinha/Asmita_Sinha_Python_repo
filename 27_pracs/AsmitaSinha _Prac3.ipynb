3. Reverse a given number

```python
print("Asmita Sinha, 24204A0027")
num =  int(input("Enter A number: "))
reverse = 0
while num > 0:
    reverse = reverse * 10 + num % 10
    num //= 10

print("Reversed Number : ", reverse)
```

4. Check whether a number is a palindrome using a loop


```python
print("Asmita Sinha, 24204A0027")
num =  int(input("Enter A number: "))
temp = num
reverse = 0

while temp > 0:
    reverse = reverse * 10 + temp % 10
    temp //= 10

if num == reverse:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

4. Check whether a given number is prime


```python
print("Asmita Sinha, 24204A0027")
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")

```
